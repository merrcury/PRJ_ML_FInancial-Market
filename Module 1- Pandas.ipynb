## Query 1.5

Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.<br>
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def avg_value(days):
    req_value_1 = df[df.Date.isin(pd.date_range(end='13/5/2019', periods=days))]\
    ['Close Price'].mean()
    print("Average for last {0} days is {1}".format(days,req_value_1))
```

```python
def pro_los_value(days):
    new_value = df[df.Date.isin(pd.date_range(end='13/5/2019', periods=days))]\
    ['Close Price'].iloc[-1]
    old_value = df[df.Date.isin(pd.date_range(end='13/5/2019', periods=days))]\
    ['Close Price'].iloc[0]
    req_value_2 = (new_value - old_value)
    req_value_2 = req_value_2/old_value
    req_value_2 = req_value_2*100
    if req_value_2 > 0:
        print("Profit Percentage over last {0} days is {1}%".format\
              (days,round(req_value_2,3)))
    elif req_value_2 < 0:
        print("Loss Percentage over last {0} days is {1}%".format\
              (days,round(abs(req_value_2),3)))
    else:
        print("No Profit - No Loss")
```

```python
while(True):
    try:
        print("\n 1.Compute Average & Profit/Loss Percentage 0. Exit")
        d = int(input("Enter Choice:  "))
        if(d == 1):
            c = int(input("Enter No of days: "))
            avg_value(c)
            pro_los_value(c)
        elif(d == 0):
            break
        else:
            print("Wrong Choice")
    except ValueError:
        print("Numbers Only")
        pass
```

## Query 1.6

Add a column 'Day_Perc_Change' where the values are the daily change in percentages i.e. the percentage change between 2 consecutive day's closing prices. Instead of using the basic mathematical formula for computing the same, use 'pct_change()' function provided by Pandas for dataframes. You will note that the first entry of the column will have a ‘Nan’ value. Why does this happen? Either remove the first row, or set the entry to 0 before proceeding.

```python
df['Day_Perc_Change'] = df['Close Price'].pct_change(periods = 1)
df['Day_Perc_Change'] = df['Day_Perc_Change']*100
```

```python
#Calculating Value for 1st Row using Prev Closing Value
#No need to drop the row or set it to zero
number1 = df.loc[df.index,'Close Price'].iat[0]
number2 = df.loc[df.index,'Prev Close'].iat[0]
number = (number1 - number2)
number = number/number2
number = number * 100
df['Day_Perc_Change'][0] = number
```

## Query 1.7

Add another column 'Trend' whose values are:

    'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
    'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
    'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
    'Positive' for 'Day_Perc_Change' in between 1 and 3
    'Negative' for 'Day_Perc_Change' in between -1 and -3

    'Among top losers' for 'Day_Perc_Change' in between -3 and -7
    'Bull run' for 'Day_Perc_Change' >7

```python
rkd = RangeKeyDict({
        (-0.5,0.5): 'Slight or No change',
        (0.5,1): 'Slight positive',
        (-1,-0.5): 'Slight negative',
        (1,3):'Positive',
        (-3,-1):'Negative',
        (3,7):'Among top gainers',
        (-7,-3):'Among top losers',
        (-10000, -7):'Bull run',
        (7, 10000 ):'Bear drop',
    })
```

```python
df["Trend"] = np.nan
for xi in range(0,494):
    df['Trend'][xi] = rkd[df['Day_Perc_Change'][xi]]
```
